```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset

titanic = pd.read_csv('Titanic-Dataset.csv')
```

```python
# Display the first few rows of the dataset
print(titanic.head())
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.head()
```

```python
# Check for missing values
print("Missing Values:")
titanic.isnull().sum()
```

```python
# Fill missing values in the 'age' column with the mean age

titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
titanic.drop('Cabin', axis=1, inplace=True)
```

```python
titanic.isnull().sum()
```

```python
# Visualize the distribution of passenger classes
sns.countplot(x='Pclass', data=titanic)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Survival rate by passenger class
sns.countplot(x='Survived', data=titanic)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=titanic)
plt.title('Survival Count by Pclass')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=titanic)
plt.title('Survival Count by Sex')
plt.show()
```

```python
sns.histplot(titanic['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.histplot(titanic['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
sns.pairplot(titanic[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
plt.suptitle('Pairplot of Variables')
plt.show()
```
